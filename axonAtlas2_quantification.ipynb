```python
#bg_heatmap(values = axon_dict, annotate_regions=False, thickness=20)

# Only keep left hemisphere regions
#left_axon_dict = {k: v for k, v in axon_dict.items() if k.endswith('_L')}

#bg_heatmap(values=left_axon_dict, annotate_regions=False, thickness=20, vmin =0, vmax=20)

# --- Split DataFrame into left and right hemisphere dictionaries ---
left_dict = {}
right_dict = {}

for _, row in df.iterrows():
    # Remove '_left' and '_right' suffix for compatibility with BrainGlobe acronyms
    if row['hemisphere'] == 'Left':
        base_acronym = row['acronym'].replace('_left', '')
        left_dict[base_acronym] = row['percentage']
    elif row['hemisphere'] == 'Right':
        base_acronym = row['acronym'].replace('_right', '')
        right_dict[base_acronym] = row['percentage']

# --- Visualize left hemisphere ---
print("Left hemisphere heatmap:")
bg_heatmap(values=left_dict, vmin=0, vmax=20, annotate_regions=False, thickness=20)

# --- Visualize right hemisphere ---
print("Right hemisphere heatmap:")
bg_heatmap(values=right_dict, vmin=0, vmax=20, annotate_regions=False, thickness=20)
```
