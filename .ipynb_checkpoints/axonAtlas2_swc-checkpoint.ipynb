```python
from brainrender import Scene
from brainrender.actors import make_neurons # Import make_neurons instead of Neuron
# import morphapi # Not needed for this code

# Create a scene
scene = Scene()

# Path to the neuron data
neuron_path = r"C:\Users\gangliagurdian\Downloads\mlnb-export-data\AA1625.swc"

# Use the factory function to create and add the neuron(s) in one step
# Note: 'make_neurons' returns a list, so we unpack it with *
neurons = scene.add(*make_neurons(neuron_path, neurite_radius=12))

# Add brain regions
scene.add_brain_region('SSp')

# Render the scene
scene.render()
```
